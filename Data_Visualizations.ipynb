### Determining Graph Titles

### Scatterplots

**Duration vs. Tempo**

```python
# drops any rows where Duration or Tempo is 0, cleans up data
msd_drop_dur = msd.loc[msd['Duration'] > 0]
msd_drop_durtemp = msd_drop_dur.loc[msd_drop_dur['Tempo'] > 0]

duration_tempo = graph_visuals(msd_drop_durtemp['Duration'], Duration, msd_drop_durtemp['Tempo'], Tempo)
duration_tempo.scatter_plot()

# name = graph_visuals(df['columnname1'], x-axis-title, df['columnname2'], y-axis-title)
# name.scatter_plot()
```

**Tempo vs. Time Signature**

```python
tempo_timesignature = graph_visuals(msd['Tempo'], Tempo, msd['TimeSignature'], TimeSignature)
tempo_timesignature.scatter_plot()
```

**Duration vs. Time Signature**

```python
# drops any rows where the seconds for Duration is longer than 700, better for viewing patterns
msd_drop_dur = msd.loc[msd['Duration'] < 700]

duration_timesignature = graph_visuals(msd_drop_dur['Duration'], Duration,msd_drop_dur['TimeSignature'], TimeSignature)
duration_timesignature.scatter_plot()
```

**Duration vs. Song Hotness**

```python
# drops any rows that have songHotness as 0
msd_droppedHot = msd.loc[msd["SongHotness"] > 0 ]

duration_songhotness = graph_visuals(msd_droppedHot['Duration'], Duration, msd_droppedHot['SongHotness'], SongHotness)
duration_songhotness.scatter_plot()
```

### Frequency Bar Graphs and Histograms
