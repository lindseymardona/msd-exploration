### Scatterplots

**Duration vs. Tempo**

```python
# drops any rows where Duration or Tempo is 0, cleans up data
msd_drop_dur = msd.loc[msd['Duration'] > 0]
msd_drop_durtemp = msd_drop_dur.loc[msd_drop_dur['Tempo'] > 0]

duration_tempo = graph_visuals(msd_drop_durtemp['Duration'], Duration, msd_drop_durtemp['Tempo'], Tempo)
duration_tempo.scatter_plot()

# name = graph_visuals(df['columnname1'], x-axis-title, df['columnname2'], y-axis-title)
# name.scatter_plot()
```

```python
fig = plt.figure(figsize =(14, 7))
plt.boxplot(msd_drop_durtemp['Tempo'],  vert = 0)
plt.show()
# create a boxplot displays Tempo data
```

```python
fig = plt.figure(figsize =(14, 7))
plt.boxplot(msd_drop_durtemp['Duration'],  vert = 0)
plt.show()
# create boxplot displaying the Duration data
```

The duration of the tracks was mostly between approximately 150 second, about two minutes and 30 seconds, and 275 seconds, about 4 minutes and 30 seconds. It can also be noted that a majority of tracks had a tempo from approximately 100 bpm to 150 bpm. The average pop song is around 116 bpm, which is right in the middle of the cluster, possibly hinting that a large number of tracks are from the pop genre.

### Frequency Bar Graphs and Histograms

**Frequency Bar Graph, Year**

```python
# Drops rows where the value for Year is 0
msd_drop_year = msd.loc[msd["Year"] > 0 ]

year_data = graph_visuals(msd_drop_year['Year'],  Year)
year_data.frequency_bar_graph()
```

```python
year_data_hist = graph_visuals(msd_drop_year['Year'], Year, bins = 30)
year_data_hist.hist_graph()
```

```python
print(msd_drop_year.shape)
print((msd_drop_year['Year']==1954).sum())
print((msd_drop_year['Year']==2006).sum())
```

We can see that the dataset is mainly composed of tracks released in the early 2000's. Out of the 4,380 tracks, after cleaning the dataset of any 'Year' values of 0, the greatest number of songs can be found in 2006. The greatest number of songs are clustered around 2003-2009, and the least number of tracks are found in the year 1954, with only one track. The year 2006 has 293 tracks within the cleaned msd dataset.

**Histogram, Tempo**

```python
# drops rows where the value of Tempo is 0
msd_drop_tempo = msd.loc[msd["Tempo"] > 0 ]

tempo_data = graph_visuals(msd_drop_tempo['Tempo'], Tempo, bins = 20)
tempo_data.hist_graph()
```

This histogram of BPM for each track better visualizes what BPM values were more common. We can see a peak in the histogram between approximatley 75 bpm and 140 bpm, reestablishing observations we made for Tempo in previous scatter plots.
